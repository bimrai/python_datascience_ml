Now focusing on discovering and how to create random distributions of data. There are many to choose from but here are a few:

```python
# just type np.random. + 'tab key', it will list many options

# first one

np.random.rand(1) # will return a number that falls between 0 and 1 
```

```python
# if you wish to pass in shape
np.random.rand(3, 4) # 3 by 4, all numbers between 0 and 1
```

```python
# in the physical real world, things fall under normal distribution. A simplified version of a normal distribution is a standard normal distribution

# np.random.randn, include n at the end of rand.

np.random.randn(4)
# this will return amount of numbers.
# Notice we will get negative numbers. Numbers that are closest to 0 have a higher rate of being selected
# than numbers further and further away from it.
```

```python
# you can also ask for these numbers in a particular shape

np.random.randn(2, 3)
```

So this is returning samples from a standard normal distribution. Because it's standard normal,  values closer to 0 are more likely to appear than values further away from 0.

```python
# here is another one
# np.random.randint(), you can tell it returns random integers

np.random.randint(0, 101, 10) # returns 0 to 101 range of random numbers, returns 10 numbers within that range
```

```python
np.random.randint(0, 101, (5, 5)) # this way it returns a 5D where 5 rows and 5 columns are generated randomly
```

When you're using random data and you run different tests on it, it's very useful to be able to set a seed. A seed is used to set a random state so that the random results can be reproduced. Results are still random, but you are choosing an arbitrary seed number, so you get a particular set of random numbers. 

```python
# to do this you do

# arbitrary number choice is 42 by default used in a lot of python libraries
np.random.seed(42)  # so this is setting the seed

# now call it, MAKE SURE ITS IN THE SAME CELL
np.random.rand(4) # after setting seed we ask for 4 random numbers between 0 and 1
```

The seed is not a container that holds the random numbers. Instead, the seed is the starting point for the math formula that creates them.

Think of it like a book:
The Seed is the Page Number.
The Random Numbers are the Sentences on that page.

If you tell NumPy to start at "Page 10" (Seed 10), and read the next 5 sentences (5 random numbers), those sentences will be the same every time you open the book to that page. The value 5 is just the count of how many numbers you want, not where they are stored.

If you set the seed to 10 and ask for 5 numbers, you get the same 5 numbers every time because you started the "math" at the same point (10).

#### Couple more useful methods and attributes you can oftenly perform on existing NumPy array 

```python
# say we have a range of 0 - 24
arr = np.arange(0, 25)
```

```python
arr # check 
```

```python
# we can reshape arr, aka the range of numbers
arr.reshape(5, 5) # here we reshape it 5 by 5
```

One common mistakes that occur is when reshaping or shaping range of numbers where it wont fit. Say we use the same 25 numbers, if we try 5 by 4, that is 20 numbers in total since 5 x 4 = 20, but that is 5 more numbers that will not fit within the shape or rows/columns correctly. 

```python
# lets check it out to prove that it won't work

arr.reshape(5, 4)
```

Here we got an error where it states, "cannot reshape array of size 25 into shape (5,4)". So keep in mind the number of rows and columsn before you squeeze in an array of numbers where it won't fit.
