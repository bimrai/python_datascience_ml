Numpy deals with 0's differently in division. It will throw some sort of warning or error if your array includes 0.

```python
arr
```

```python
arr / arr
```

As you can see, we got that warning message: "RuntimeWarning: invalid value encountered in divide
  arr / arr".

It will still output the value regardless.

When you do this in python you get returned a ZeroDivisionError, but in mumpy you will get a warning and an output, as well as the division by 0 being returned as infinite.

```python
1 / 0 # regular python will throw out an error
```

```python
1 / arr # division with numpy array where first value output is inf for infinite where 0 was, where 1 / 0. 
```

The risk with this is, is that regardless of the problem, NumPy will keep on running the program despite the issues occuring.

#### Let's create a 2D (2 Dimensional) Array

```python
arr2d = np.arange(0, 25).reshape(5, 5)
```

```python
arr2d.shape
```

```python
arr2d
```

```python
arr2d.sum()
```

When you shift + tab on say sum(), you can read "a.sum(axis=None, dtype=None, out=None, keepdims=False, initial=0, where=True)".

You notice that sum actually takes in an axes parameter, which means you can specify whether you want to perform this operation across the rows or across the columns. Key word being across.

```python
arr2d.sum(axis=0)
```

So on axis=0, we get our first value being 50. This is the sum of the first column across the rows. We go from first row + row 2 + row 3 ... until the last row, sum all the values of which we said is 50. We do this across all columns on axis 0. 

```python
arr2d.sum(axis=1)
```

When axis=1, we can see this time, we go across all columns, not rows. We go across all columns per row and get the sum. We can see in our first row, if we sum the values across all columns again from the first row, we get 10. 4 + 3 + 2 + 1 = 10. 

# KEY TAKEAWAY
## Axis = 0: Across the rows
## Axis = 1: Across the columns
