### Broadcasting

NumPy arrays can broadcast, meaning they can take a single value and automatically apply it to many elements at once. In normal Python lists, if you want to replace multiple elements, you must provide the same number of new elements. With NumPy, you can assign just one value and NumPy will automatically repeat it across the selected part of the array.

```python
# here is an example
# so we have our array 0 - 10
arr
```

```python
arr[:5] = 100
```

Here we assign the first 0 - 5 indexed elements to equal 100, so we broadcast 100 over those selected elements.

```python
arr
```

```python
# lets just reset this again

arr = np.arange(0, 11)
arr
```

When you slice a NumPy array and store it in another variable, NumPy usually does NOT create a new array. Instead, it creates a view (a pointer) to the same original data in memory.

When you slice a NumPy array and assign it to a new variable, that new variable points to the same data in memory. Because of that, changing the slice also changes the original array.

In simple terms:
The slice is not a separate copy, it’s just another view of the original array, so any modification to it will affect the original array as well.

```python
# here is an example

arr_slice = arr[0:5] # slicing array from 0 - 4 values
```

```python
# lets check our sliced array

arr_slice
```

```python
# now lets say we select everything in that slice from the original array held in "arr", and broadcast their values to be 117 (for the halo fans). 
# that should update not only for this arr_slice, but the original array variable held in "arr".
arr_slice[:] = 117
```

```python
# now lets check our arr_slice

arr_slice
```

```python
# since that slice has been updated, lets see our original array, "arr".
arr
```

So as you can see, the NumPy array has been sliced, updated the slice to 117 from 0 to 4, which was then applied to the original array where we can see the output has changed. Key thing again, it is just a pointer, not a copy or any new created array. Just a referencing point which is actually the original. 

Moreover, broadcasting saves memory over the original Pythons method when updating the elements.

### What if you didn't want to affect the original array?

Well, you will need to explicitly set a copy of the array.

```python
arr_copy = arr.copy() # so this creates the copy of an array, duplicating the original which also means more memory usage
```

```python
arr_copy
```

```python
# to test that the original remains untouched when changes are being made, lets make all elements 100.
arr_copy[:] = 100
```

```python
# lets check that it updated in arr_copy
arr_copy
```

```python
# okay now that its updated, lets check our original array
arr
```

#### So you can see, when you set a copy and reassign it into a new variable, the original will not be touched. 

#### HOWEVER, if you reassign it into a new variable BUT YOU DO NOT SET COPY, then the original array will be affected to manipulation. 

### Indexing on 2D arrays/matrices

```python
array_2d = np.array([[5, 10, 15], [20, 25, 30], [35, 40, 46]])
# REMINDER: remember, this is essentially a list of 3 lists (nested list), which is transformed into a numpy array via np.array 
```

```python
array_2d
```

```python
array_2d.shape # returns number of (row, columns)
```

```python
# lets say we just want to grab a single row

array_2d[0] # so we are using index to grab the rows, so index 0 holds row 1.
```

```python
# if we want row 3, we choose index 2

array_2d[2]
```

```python
# say you want to grab 25. thats in row index 1, column index 1. theres 2 ways you can do this
array_2d[1][1] # method 1
```

```python
# this method is MUCH preferred over the previous one
array_2d[1, 1] # basically means [row 1, column 1]
```

```python
# we can also use slicing. say we want to get 10 and 15 from the first row, then 25 and 30 from the second row.

array_2d[:2] # first we grap the rows we want, we omit 0 and up to but not including 2. this grabs first 2 rows
```

We got our sliced target selection. So now we separate with a comma after 2 from when we sliced, then add 1 since that is the start of the column where 10 is (which will ignore the first column of 5 and 20 from first and second row) and omit the end value with just colon to go till the end since we need both 15 and 30 which are the end of the columns.

```python
array_2d[:2, 1:] 
```

## Conditional Selection

This selection is much more commonly used.

```python
# example

arr = np.arange(1, 11)
```

```python
arr
```

```python
# you can broadcast a condition, like wheres my array greater than 4?

arr > 4 # this will output a boolean result where elements that are greater than 4 will return true, else false. so that is any number 4 or below
```

```python
bool_arr = arr > 4 # assign the condition
```

```python
bool_arr
```

```python
arr[bool_arr] # you can then filter by passing it to the original array
```

```python
arr[arr > 4] # you can also filter by passing condition through original array too.
```

### arr[arr > 4] <-- This is the conditional selection or sort of thing we will see quite often, especially when you progress towards Pandas. Pandas will then later give us the ability to do multiple comparisons.
